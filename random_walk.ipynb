```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random as r

r.seed(2)

def random_walk(n):
    x = np.arange(n + 1)
    # y = np.array([random.choice([1,-1]) for i in range(0, n + 1)]).cumsum()
    y = np.array([r.choice([0, 1, 1, 1, -1,-2]) 
                  for i in range(0, n + 1)]).cumsum()
    return x, y

x,y = random_walk(1000)
```

```python
def sma(x, n):
    return pd.Series(x).rolling(n).mean().values

sma50 = sma(y, 50)
sma100 = sma(y, 100)
```

```python
bfig=plt.figure(figsize=(12,5), dpi= 100, facecolor='w', edgecolor='k')
plt.plot(x, y, '-')
plt.plot(x, sma50, '-')
plt.plot(x, sma100, '-')
# plt.title("1D Random Walk")
plt.show()
```

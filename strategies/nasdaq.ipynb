```python
import pandas as pd

# pretty printing of pandas dataframe
pd.set_option('expand_frame_repr', False)
pd.set_option("display.max_rows", None, "display.max_columns", None)

# There are 4 tables on the Wikipedia page
# we want the last table

payload=pd.read_html('https://en.wikipedia.org/wiki/Nasdaq-100')
first_table  = payload[0]
second_table = payload[1]
third_table  = payload[2]
fourth_table = payload[3]

df = fourth_table
tickers = df['Ticker'].values
len(tickers)
```

```python
print('Name,Price,Spread,RollLong,RollShort,PIP,PIPCost,MarginCost,Leverage,LotAmount,Commission,Symbol')

for t in tickers:
    print(f'{t},50,0.1,0,0,0.01,0.01,0,2,1,0.01,*')
```
